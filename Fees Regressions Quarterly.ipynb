# Fees Regressions - Quarterly

## Loading in Data

```python
import pandas as pd
import numpy as np
```

```python
excel_file_path = 'Quarterly_simplified.xlsx'

all_dfs = pd.read_excel(excel_file_path, sheet_name=None)
```

```python
print("Worksheet names:", list(all_dfs.keys()))
```

```python
fees = all_dfs['MySuper_Fees']
performance = all_dfs['Investment_Performance']
profile = all_dfs ['Fund_Profile']
```

```python
unique_mysuper_name = profile['MySuper product name'].nunique()
unique_mysuper_type = profile['Fund name'].nunique()

print("Number of unique values in 'MySuper product name':", unique_mysuper_name)
print("Number of unique values in 'MySuper product type':", unique_mysuper_type)
```

```python
profile
```

```python
# Assuming you have two dataframes: fees and performance
merged_df = fees.merge(performance, on=['Period', 'MySuper product name', 'MySuper product type', 'Fund name', 'Fund ABN'], how='inner')

returns_df = merged_df = profile.merge(merged_df, on=['Period', 'MySuper product name', 'MySuper product type', 'Fund name', 'Fund ABN'], how='inner')
```

```python
returns_df
```

## Loading in merger list

```python
excel_file_path = 'Merged SF Annual Data.xlsx'

merger_file = pd.read_excel(excel_file_path, sheet_name=None)
```

```python
print("Worksheet names:", list(merger_file.keys()))
```

```python
merger = merger_file['Quarterly']
```

```python
merger
```

```python
quarterly = pd.merge(merged_df, merger, on=['Period', 'MySuper product name', 'Fund name', 'Fund ABN'], how='left')

quarterly['Merger'] = quarterly['Merger'].fillna(0)

quarterly['Merged'] = quarterly['Merged'].fillna(0)
```

```python
returns = pd.merge(returns_df, merger, on=['Period', 'MySuper product name', 'Fund name', 'Fund ABN'], how='left')

returns['Merger'] = returns['Merger'].fillna(0)

returns['Merged'] = returns['Merged'].fillna(0)
```

## Regressions for Quarterly Data

```python
quarterly.columns
```

```python
#Investment Fees Regression - Rep Member

import statsmodels.api as sm

# Extract dependent and independent variables from df_quarterly
Y = quarterly['InvFee']
X = quarterly[['Merger', 'IndCost','onenret']]
X = sm.add_constant(X)  # Add a constant to the model (i.e., the intercept)

# Set up and run the regression model
model = sm.OLS(Y, X).fit()

# Display the regression results
print(model.summary())
```
