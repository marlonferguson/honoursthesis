```python

from IPython.core.display import display, HTML

# Assuming investment_performance_merged and investment_non_merged are your DataFrames

# List of columns to plot
columns_to_plot = ['Net investment return\n', 'Net return \n (rep member) - Quarterly', 'One-year net return \n (rep member) - Annualised']

# Loop through each column to plot
for col in columns_to_plot:
    # Calculate the average return per period for merged and non_merged DataFrames
    avg_return_merged = investment_performance_merged.groupby('Period')[col].mean().reset_index()
    avg_return_non_merged = investment_non_merged.groupby('Period')[col].mean().reset_index()
    
    # Merge the two DataFrames on 'Period'
    merged_df = pd.merge(avg_return_merged, avg_return_non_merged, on='Period', suffixes=('_merged', '_non_merged'))
    
    # Create HTML table
    html_table = "<table border='1'>"
    html_table += "<tr><th>Period</th><th>Average Return (Merged)</th><th>Average Return (Non-Merged)</th></tr>"
    
    for i, row in merged_df.iterrows():
        period = row['Period'].strftime('%d-%m-%Y')  # Format date as DD-MM-YYYY
        avg_merged = row[col + '_merged']
        avg_non_merged = row[col + '_non_merged']
        
        # Determine which value is greater and should be highlighted
        if avg_merged > avg_non_merged:
            avg_merged_str = f"<b style='color:green;'>{avg_merged:.2f}</b>"
            avg_non_merged_str = f"{avg_non_merged:.2f}"
        else:
            avg_merged_str = f"{avg_merged:.2f}"
            avg_non_merged_str = f"<b style='color:green;'>{avg_non_merged:.2f}</b>"
        
        html_table += f"<tr><td>{period}</td><td>{avg_merged_str}</td><td>{avg_non_merged_str}</td></tr>"
    
    html_table += "</table>"
    
    # Display the table
    display(HTML(f"<h2>Average {col} Over Time</h2>"))
    display(HTML(html_table))
```

```python
# List of columns to plot
columns_to_plot = ['Net investment return\n', 'Net return \n (rep member) - Quarterly', 'One-year net return \n (rep member) - Annualised']

# Initialize an empty DataFrame to store the final results
final_df = pd.DataFrame()
final_df['Period'] = investment_performance_merged['Period'].drop_duplicates().sort_values().reset_index(drop=True)

# Loop through each column to collect data
for col in columns_to_plot:
    # Calculate the average return per period for merged and non_merged DataFrames
    avg_return_merged = investment_performance_merged.groupby('Period')[col].mean().reset_index()
    avg_return_non_merged = investment_non_merged.groupby('Period')[col].mean().reset_index()
    
    # Merge these average returns into the final DataFrame
    final_df = pd.merge(final_df, avg_return_merged, on='Period', how='left')
    final_df = pd.merge(final_df, avg_return_non_merged, on='Period', how='left', suffixes=('_merged', '_non_merged'))

# Drop any rows with NaN values
final_df.dropna(inplace=True)

# Create HTML table
html_table = "<table border='1' style='width:100%;'>"
html_table += "<tr><th rowspan='2'>Period</th><th colspan='2'>Net Investment Return</th><th colspan='2'>Net Return (Rep Member) - Quarterly</th><th colspan='2'>One-Year Net Return (Rep Member) - Annualised</th></tr>"
html_table += "<tr><th>Merged</th><th>Non-Merged</th><th>Merged</th><th>Non-Merged</th><th>Merged</th><th>Non-Merged</th></tr>"

for i, row in final_df.iterrows():
    period = row['Period'].strftime('%d-%m-%Y')  # Format date as DD-MM-YYYY
    
    html_table += f"<tr><td>{period}</td>"
    
    for col in columns_to_plot:
        avg_merged = row[col + '_merged']
        avg_non_merged = row[col + '_non_merged']
        
        # Determine which value is greater and should be highlighted
        if avg_merged > avg_non_merged:
            avg_merged_str = f"<b style='color:green;'>{avg_merged:.2f}</b>"
            avg_non_merged_str = f"{avg_non_merged:.2f}"
        else:
            avg_merged_str = f"{avg_merged:.2f}"
            avg_non_merged_str = f"<b style='color:green;'>{avg_non_merged:.2f}</b>"
        
        html_table += f"<td>{avg_merged_str}</td><td>{avg_non_merged_str}</td>"
    
    html_table += "</tr>"

html_table += "</table>"

# Display the table
display(HTML(html_table))
```

```python
# List of columns to plot
columns_to_plot = ['Net investment return\n', 'Net return \n (rep member) - Quarterly', 'One-year net return \n (rep member) - Annualised']

# Initialize an empty DataFrame to store the differences
difference_df = pd.DataFrame()

# Loop through each column to calculate the differences
for col in columns_to_plot:
    # Calculate the average return per period for merged and non_merged DataFrames
    avg_return_merged = investment_performance_merged.groupby('Period')[col].mean().reset_index()
    avg_return_non_merged = investment_non_merged.groupby('Period')[col].mean().reset_index()
    
    # Merge the two DataFrames on 'Period' to make sure we compare the same periods
    merged_df = pd.merge(avg_return_merged, avg_return_non_merged, on='Period', suffixes=('_merged', '_non_merged'))
    
    # Calculate the difference between merged and non-merged returns for each period
    merged_df['Difference'] = merged_df[col + '_merged'] - merged_df[col + '_non_merged']
    
    # Add a new column to indicate which metric the difference corresponds to
    merged_df['Metric'] = col
    
    # Append the differences to the main DataFrame
    difference_df = pd.concat([difference_df, merged_df[['Difference', 'Metric']]])

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Metric', y='Difference', data=difference_df)
plt.title('Box Plot of Differences in Returns')
plt.xlabel('Metric')
plt.ylabel('Difference in Return (Merged - Non-Merged)')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for col in columns_to_plot:
    sns.histplot(difference_df[difference_df['Metric'] == col]['Difference'], label=col, kde=True)
plt.title('Histogram of Differences in Returns')
plt.xlabel('Difference in Return')
plt.ylabel('Frequency')
plt.legend(title='Metric')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Metric', y='Difference', data=difference_df)
plt.title('Violin Plot of Differences in Returns')
plt.xlabel('Metric')
plt.ylabel('Difference in Return')
plt.xticks(rotation=45)
plt.show()
```
