```python
# Function to replace NaN, blank, or '*' with 0 for specific columns
def replace_values(df):
    cols_to_exclude = ['Period', 'Fund name', 'ABN', 'Acquirer', 'Merged', 'Merger', 'Fund type']
    cols_to_include = [col for col in df.columns if col not in cols_to_exclude]
    
    # Debugging
    print("Number of cols in original df:", len(df.columns))
    print("Number of cols to include:", len(cols_to_include))

    # Replacing the specific values with 0 for the included columns
    df[cols_to_include] = df[cols_to_include].replace(['*', ''], 0).fillna(0)
    return df

# Update the values for specific DataFrames
fees = replace_values(fees)
fees_merged = replace_values(fees_merged)
fees_non_merged = replace_values(fees_non_merged)

financial_performance = replace_values(financial_performance)
financial_merged = replace_values(financial_merged)
financial_non_merged = replace_values(financial_non_merged)

fund_profile = replace_values(fund_profile)
fund_profile_merged = replace_values(fund_profile_merged)
fund_profile_non_merged = replace_values(fund_profile_non_merged)
```

```python
# Function to convert columns to desired data types
def convert_columns(df):
    for col in df.columns:
        if col == 'Period':
            df[col] = pd.to_datetime(df[col])
        elif col not in ['Fund name', 'Acquirer', 'Fund type']:
            df[col] = df[col].astype('float64')
    return df

# Update the data types for each DataFrame
fees = convert_columns(fees)
fees_merged = convert_columns(fees_merged)
fees_non_merged = convert_columns(fees_non_merged)
financial_performance = convert_columns(financial_performance)
financial_merged = convert_columns(financial_merged)
financial_non_merged = convert_columns(financial_non_merged)
fund_profile = convert_columns(fund_profile)
fund_profile_merged = convert_columns(fund_profile_merged)
fund_profile_non_merged = convert_columns(fund_profile_non_merged)
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine
with pd.ExcelWriter('Annual_simplified.xlsx', engine='xlsxwriter') as writer:
    
    # Write each DataFrame to a specific sheet
    fees.to_excel(writer, sheet_name='Fees', index=False)
    fees_merged.to_excel(writer, sheet_name='Fees_Merged', index=False)
    fees_non_merged.to_excel(writer, sheet_name='Fees_Non_Merged', index=False)
    financial_performance.to_excel(writer, sheet_name='Financial_Performance', index=False)
    financial_merged.to_excel(writer, sheet_name='Financial_Merged', index=False)
    financial_non_merged.to_excel(writer, sheet_name='Financial_Non_Merged', index=False)
    fund_profile.to_excel(writer, sheet_name='Fund_Profile', index=False)
    fund_profile_merged.to_excel(writer, sheet_name='Fund_Profile_Merged', index=False)
    fund_profile_non_merged.to_excel(writer, sheet_name='Fund_Profile_Non_Merged', index=False)
```

```python
# Function to replace NaN, blank, or '*' with 0 for specific columns
def replace_values(df):
    cols_to_exclude = ['Period', 'Fund name', 'Acquirer', 'Merged', 'Merger', 'Fund type']
    cols_to_include = [col for col in df.columns if col not in cols_to_exclude]
    df[cols_to_include] = df[cols_to_include].replace(['*', ''], 0).fillna(0)
    return df

# Renaming columns before replacing values
fund_profile.rename(columns=fund_profile_new_columns, inplace=True)
fund_profile_merged.rename(columns=fund_profile_merged_new_columns, inplace=True)
fund_profile_non_merged.rename(columns=fund_profile_non_merged_new_columns, inplace=True)

investment_performance.rename(columns=investment_performance_new_columns, inplace=True)
investment_performance_merged.rename(columns=investment_performance_merged_new_columns, inplace=True)
investment_non_merged.rename(columns=investment_performance_non_merged_new_columns, inplace=True)

mysuper_fees.rename(columns=mysuper_fees_new_columns, inplace=True)
mysuper_fees_non_merged.rename(columns=mysuper_fees_non_merged_new_columns, inplace=True)
mysuper_fees_merged.rename(columns=mysuper_fees_merged_new_columns, inplace=True)


admin_fees.rename(columns=admin_fees_new_columns, inplace=True)
admin_fees_non_merged.rename(columns=admin_fees_non_merged_new_columns, inplace=True)
admin_fees_merged.rename(columns=admin_fees_merged_new_columns, inplace=True)

activity_fees.rename(columns=activity_fees_new_columns, inplace=True)
activity_fees_non_merged.rename(columns=activity_fees_non_merged_new_columns, inplace=True)
activity_fees_merged.rename(columns=activity_fees_merged_new_columns, inplace=True)

# Update the values for new DataFrames
fund_profile = replace_values(fund_profile)
fund_profile_merged = replace_values(fund_profile_merged)
fund_profile_non_merged = replace_values(fund_profile_non_merged)

investment_performance = replace_values(investment_performance)
investment_performance_merged = replace_values(investment_performance_merged)
investment_non_merged = replace_values(investment_non_merged)

mysuper_fees = replace_values(mysuper_fees)
mysuper_fees_merged = replace_values(mysuper_fees_merged)
mysuper_fees_non_merged = replace_values(mysuper_fees_non_merged)

admin_fees = replace_values(admin_fees)
admin_fees_merged = replace_values(admin_fees_merged)
admin_fees_non_merged = replace_values(admin_fees_non_merged)

activity_fees = replace_values(activity_fees)
activity_fees_merged = replace_values(activity_fees_merged)
activity_fees_non_merged = replace_values(activity_fees_non_merged)
```

```python
# Function to convert columns to desired data types
def convert_columns(df):
    for col in df.columns:
        if col == 'Period':
            df[col] = pd.to_datetime(df[col])
        elif col not in ['MySuper product name', 'MySuper product type', 'Fund name','Fund ABN', 'Merger', 'Merged', 'Acquirer']:
            df[col] = df[col].astype('float64')
    return df

# Update the data types for new DataFrames
fund_profile = convert_columns(fund_profile)
fund_profile_merged = convert_columns(fund_profile_merged)
fund_profile_non_merged = convert_columns(fund_profile_non_merged)
investment_performance = convert_columns(investment_performance)
investment_performance_merged = convert_columns(investment_performance_merged)
investment_non_merged = convert_columns(investment_non_merged)
mysuper_fees = convert_columns(mysuper_fees)
mysuper_fees_merged = convert_columns(mysuper_fees_merged)
mysuper_fees_non_merged = convert_columns(mysuper_fees_non_merged)
admin_fees = convert_columns(admin_fees)
admin_fees_merged = convert_columns(admin_fees_merged)
admin_fees_non_merged = convert_columns(admin_fees_non_merged)
activity_fees = convert_columns(activity_fees)
activity_fees_merged = convert_columns(activity_fees_merged)
activity_fees_non_merged = convert_columns(activity_fees_non_merged)
```
