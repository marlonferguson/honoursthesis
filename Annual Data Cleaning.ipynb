**Creating sub-sets for firms that have merged and firms that have not merged**

```python
# Convert the 'Period' column in 'master_list' dataframe to datetime
master_list['Period'] = pd.to_datetime(master_list['Period'])
```

```python
# Convert the 'Period' column to datetime
financial_performance['Period'] = pd.to_datetime(financial_performance['Period'])

# Merge the dataframe based on 'Period', 'Fund name', 'ABN', and 'MySuper Product Name'
financial_performance_merged = pd.merge(master_list, financial_performance, on=['Period', 'Fund name', 'ABN'], how='inner')

# Create a key column in both DataFrames for easier comparison
financial_performance['key'] = financial_performance['Period'].astype(str) + financial_performance['Fund name'].astype(str) + financial_performance['ABN'].astype(str)
financial_performance_merged['key'] = financial_performance_merged['Period'].astype(str) + financial_performance_merged['Fund name'].astype(str) + financial_performance_merged['ABN'].astype(str)

# Identify non-merged funds
financial_performance_non_merged = financial_performance[~financial_performance['key'].isin(financial_performance_merged['key'])]

# Drop the key column as it's no longer needed
financial_performance.drop(columns=['key'], inplace=True)
financial_performance_merged.drop(columns=['key'], inplace=True)
financial_performance_non_merged.drop(columns=['key'], inplace=True)
```

```python
# Convert the 'Period' column to datetime
fees['Period'] = pd.to_datetime(fees['Period'])

# Merge the dataframe based on 'Period', 'Fund name', 'ABN', and 'MySuper Product Name'
fees_merged = pd.merge(master_list, fees, on=['Period', 'Fund name', 'ABN'], how='inner')

# Create a key column in both DataFrames for easier comparison
fees['key'] = fees['Period'].astype(str) + fees['Fund name'].astype(str) + fees['ABN'].astype(str)
fees_merged['key'] = fees_merged['Period'].astype(str) + fees_merged['Fund name'].astype(str) + fees_merged['ABN'].astype(str)

# Identify non-merged funds
fees_non_merged = fees[~fees['key'].isin(fees_merged['key'])]

# Drop the key column as it's no longer needed
fees.drop(columns=['key'], inplace=True)
fees_merged.drop(columns=['key'], inplace=True)
fees_non_merged.drop(columns=['key'], inplace=True)
```

```python
# Convert the 'Period' column to datetime
fund_profile['Period'] = pd.to_datetime(fund_profile['Period'])

# Merge the dataframe based on 'Period', 'Fund name', 'ABN', and 'MySuper Product Name'
fund_profile_merged = pd.merge(master_list, fund_profile, on=['Period', 'Fund name', 'ABN'], how='inner')

# Create a key column in both DataFrames for easier comparison
fund_profile['key'] = fund_profile['Period'].astype(str) + fund_profile['Fund name'].astype(str) + fund_profile['ABN'].astype(str)
fund_profile_merged['key'] = fund_profile_merged['Period'].astype(str) + fund_profile_merged['Fund name'].astype(str) + fund_profile_merged['ABN'].astype(str)

# Identify non-merged funds
fund_profile_non_merged = fund_profile[~fund_profile['key'].isin(fund_profile_merged['key'])]

# Drop the key column as it's no longer needed
fund_profile.drop(columns=['key'], inplace=True)
fund_profile_merged.drop(columns=['key'], inplace=True)
fund_profile_non_merged.drop(columns=['key'], inplace=True)
```

Confirming Merger has worked properly

```python
# Count missing values (NAT) in the "Period" column
missing_count = fund_profile_non_merged['Period'].isnull().sum()

print("Number of missing values (NAT) in the 'Period' column:", missing_count)
```

```python
# Get unique fund names in the 'Fund Name' column for both dataframes
unique_names_merged = set(fees_merged['Fund name'].unique())
unique_names_non_merged = set(fees_non_merged['Fund name'].unique())

# Find the common fund names between the two dataframes
common_names = unique_names_merged.intersection(unique_names_non_merged)

# Print the common fund names
print("Funds that appear in both fees_merged and fees_non_merged:")
for name in common_names:
    print(name)
```

**Using RSE wind-up date as merged fund dummy variable**

```python
# Convert the 'RSE Wind up date' column to datetime
financial_performance_merged['RSE Wind up date'] = pd.to_datetime(financial_performance_merged['RSE Wind up date'], errors='coerce')

# Group by 'Fund Name' and aggregate the non-NaN 'RSE Wind up date' column
unique_funds_rse = financial_performance_merged.groupby('Fund name')['RSE Wind up date'].apply(lambda x: x.dropna().dt.strftime('%Y/%m/%d').unique())

# Print the unique fund names and their corresponding RSE Wind up dates
for fund_name, rse_dates in unique_funds_rse.items():
    if rse_dates.size > 0:
        print(f"Fund Name: {fund_name}")
        print(f"RSE Wind up dates: {', '.join(rse_dates)}")
        print("---")
```

```python
# Create an empty list to store the matched RSE Wind up dates
matched_rse_dates = []

# Iterate through each row in merger_list
for _, row in merger_list.iterrows():
    acquired_fund = row['Acquired Fund']
    if acquired_fund in unique_funds_rse:
        rse_dates = unique_funds_rse[acquired_fund]
        matched_rse_dates.append(', '.join(rse_dates))
    else:
        matched_rse_dates.append('No Match')

# Add the matched RSE Wind up dates as a new column in merger_list
merger_list['Matched RSE Wind up dates'] = matched_rse_dates

# Print the modified merger_list dataframe
print(merger_list)
```

Checking Mergers

```python
dataframes = {
    'fees': fees,
    'fees_merged': fees_merged,
    'fees_non_merged': fees_non_merged,
    'financial_performance': financial_performance,
    'financial_performance_merged': financial_performance_merged,
    'financial_performance_non_merged': financial_performance_non_merged,
    'fund_profile': fund_profile,
    'fund_profile_merged': fund_profile_merged,
    'fund_profile_non_merged': fund_profile_non_merged,
}

# Create a master list of periods (you can customize this based on your specific needs)
master_periods = set()  # Initialize an empty set to store unique periods
for df_name, df in dataframes.items():
    if 'Period' in df.columns:
        master_periods.update(df['Period'].dt.strftime('%Y-%m-%d').unique())  # Add unique periods from each DataFrame to the master set

# Check for missing periods in each DataFrame
for df_name, df in dataframes.items():
    if 'Period' in df.columns:
        df_periods = set(df['Period'].dt.strftime('%Y-%m-%d').unique())
        missing_periods = master_periods - df_periods  # Find periods that are in the master list but not in the current DataFrame
        if missing_periods:
            print(f"Missing periods in {df_name}: {sorted(list(missing_periods))}")
        print('-' * 50)
```
